```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("telecom_churn.csv")
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Churn', y='CustServCalls', data=df, estimator='mean')
plt.title('Average Customer Service Calls by Churn Status')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Avg Cust Service Calls')
plt.savefig('bar_churn_calls.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='DataPlan', y='MonthlyCharge', data=df, estimator='mean')
plt.title('Average Monthly Charge by Data Plan')
plt.xlabel('Data Plan (0 = No, 1 = Yes)')
plt.ylabel('Avg Monthly Charge')
plt.savefig('bar_dataplan_charge.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='DataUsage', y='MonthlyCharge', hue='Churn', data=df, alpha=0.6)
plt.title('Data Usage vs Monthly Charge')
plt.xlabel('Data Usage (GB)')
plt.ylabel('Monthly Charge')
plt.savefig('scatter_usage_charge.png')
plt.show()
```

```python
avg_overage_by_week = df.groupby('AccountWeeks')['OverageFee'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='AccountWeeks', y='OverageFee', data=avg_overage_by_week, color='blue')
plt.title('Average Overage Fee over Account Weeks')
plt.xlabel('Account Weeks')
plt.ylabel('Avg Overage Fee')
plt.savefig('line_overage_weeks.png')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Telecom Data')
plt.savefig('heatmap_correlation.png')
plt.show()
```
